```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_dummy = np.random.rand(100, 1)
y_dummy = 2 * x_dummy + 1 + 0.1 * np.random.rand(100, 1)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])
model.compile(optimizer='sgd', loss='mse')
model.fit(x_dummy, y_dummy, epochs=100)
predictions = model.predict(x_dummy)
plt.scatter(x_dummy, y_dummy, label='Data')
plt.plot(x_dummy, predictions, color='red', label='Model Prediction')
plt.legend()
plt.show()
```
